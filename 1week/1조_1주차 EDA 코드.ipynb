### 데이터 전처리

```python
# # Drop rows with zeros in any column except for X3, X5, and Y
# columns_to_check = data.columns.difference(['X3', 'X5', 'Y'])
# data = data[(data[columns_to_check] != 0).all(axis=1)]
```

```python
#상관계수가 높은 X17,18,19,20 drop
data = data.drop(columns=['X17', 'X18', 'X19', 'X20'])
```

```python
# IQR 방법을 사용하여 이상치 제거
def remove_outliers(data, columns):
    for col in columns:
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        data = data[(data[col] >= lower_bound) & (data[col] <= upper_bound)]
    return data
```

```python
columns_to_check = data.columns.difference(['Y'])
df = remove_outliers(data, columns_to_check)
```
